```python
import numpy as np
import matplotlib.pyplot as plt
import nltk
import re
```

```python
a = open('trump.txt', 'r')
trump_speech = a.read()
a.close()
```

```python
print("Part of Trump Speech: \n")
print(trump_speech[:995])
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
sentences = sent_tokenize(trump_speech)
```

```python
trump_sentences = []
for i in sentences:
    i = i.lower()
    i = re.sub(",", "", i)
    i = re.sub('"', "", i)
    i = re.sub("'", "", i)
    i = re.sub('’', "", i)
    i = re.sub('—', "", i)
    i = "{" + i
    if(i[-1] == "."):
        i = i[:-1]+"}"
    else:
        i = i[:-1]+"}"
#     i = re.sub(".", "", i)
    i = word_tokenize(i)
    if("SPEECH" not in i):
        trump_sentences.append(i)       
```

```python
import random
random.shuffle(trump_sentences)
train_sentences = trump_sentences[:int(len(trump_sentences)*0.8)]
test_sentences = trump_sentences[int(len(trump_sentences)*0.8):]
```

## Unigram Model

```python
unigram_model = {}
for i in train_sentences:
    for j in i:
        if(str((j, )) in unigram_model):
            unigram_model[str((j, ))] += 1
        else:
            unigram_model[str((j, ))] = 1

# total = sum(unigram_model.values())
# for i in unigram_model:
#     unigram_model[i] = unigram_model[i]/total
# print(unigram_model)
```

## Bigram Model

```python
from nltk import bigrams
bigram_model = {}

for sentence in train_sentences:
    for w1, w2 in bigrams(sentence):
        if(str((w1, )) in bigram_model):
            if(w2 in bigram_model[str((w1, ))]):
                bigram_model[str((w1, ))][w2] += 1
            else:
                bigram_model[str((w1, ))][w2] = 1
        else:
            bigram_model[str((w1, ))] = {}
            bigram_model[str((w1, ))][w2] = 1
 
# for i in bigram_model:
#     total = sum(bigram_model[i].values())
#     for j in bigram_model[i]:
#         bigram_model[i][j] = bigram_model[i][j]/total
```

## Trigram Model

```python
from nltk import trigrams
trigram_model = {}

for sentence in train_sentences:
    for w1, w2, w3 in trigrams(sentence):
        if(str((w1, w2)) in trigram_model):
            if(w3 in trigram_model[str((w1, w2))]):
                trigram_model[str((w1, w2))][w3] += 1
            else:
                trigram_model[str((w1, w2))][w3] = 1
        else:
            trigram_model[str((w1, w2))] = {}
            trigram_model[str((w1, w2))][w3] = 1
 
# for i in trigram_model:
#     total = sum(trigram_model[i].values())
#     for j in trigram_model[i]:
#         trigram_model[i][j] = trigram_model[i][j]/total
```

## quadgram Model

```python
from nltk import ngrams
quadgram_model = {}

for sentence in train_sentences:
    for w1, w2, w3, w4 in ngrams(sentence, 4):
        if(str((w1, w2, w3)) in quadgram_model):
            if(w4 in quadgram_model[str((w1, w2, w3))]):
                quadgram_model[str((w1, w2, w3))][w4] += 1
            else:
                quadgram_model[str((w1, w2, w3))][w4] = 1
        else:
            quadgram_model[str((w1, w2, w3))] = {}
            quadgram_model[str((w1, w2, w3))][w4] = 1
 
# for i in quadgram_model:
#     total = sum(quadgram_model[i].values())
#     for j in quadgram_model[i]:
#         quadgram_model[i][j] = quadgram_model[i][j]/total
```

```python
import math
unigram_MLE = []
unigram_total = sum(unigram_model.values())
for i in train_sentences:
    mle = 1
    for j in i:
        if(str((j,)) in unigram_model.keys()):
            mle = mle + math.log(unigram_model[str((j,))]/unigram_total)
        else:
            mle = 0
            break
    unigram_MLE.append(mle)
```

```python
bigram_MLE = []
for i in train_sentences:
    mle = 1
    for j in range(len(i)-1):
        if(str((i[j], )) in bigram_model.keys()):
            if(i[j+1] in bigram_model[str((i[j], ))].keys()):
                mle = mle + math.log(bigram_model[str((i[j], ))][i[j+1]]/sum(bigram_model[str((i[j], ))].values()))
            else:
                mle = 0
                break
        else:
            mle = 0
            break
    bigram_MLE.append(mle)
```
